```python
# Now we can make and clean dataframes to answer the questions in the assignment.
#I start with creating a new dataframe with the relevant columns in it, and changing 'species ' into 'species'
df.rename(columns={"Species ": "Species"}, inplace=True)
df_species = df[['Fatal (Y/N)', 'Species']]
df_species
```

```python
#Lets start by looking at the species column. This column really needs to be cleaned. We can see what the data looks like by running some diagnotics (module 5)
def print_separator(sep, num, msg):
  print("\n")
  print(sep * num)
  print(f"{msg}")
  print(sep * num)


def look_at_unique_values(column):
  unique_values_cutoff = 160
  unique_values = column.unique()
  num_unique_values = len(unique_values)
  if num_unique_values == len(column):
    print(f"Each value in the column is unique (total: {num_unique_values})")
  elif num_unique_values < unique_values_cutoff:
    print(f"Less than {unique_values_cutoff} unique values:")
    # We may get an error when sorting
    try:
      sorted = np.sort(unique_values)
      print("Values are sorted")
      display(list(sorted))
    except:
      print("Could not sort values")
      display(list(unique_values))
  else:
    print(f"More than {unique_values_cutoff} unique values (total: {num_unique_values})")


def look_at_edges(df, column_name):
  # inner function
  def show_head_and_tail(values):
      num_items_to_slice = 10
      display(list(values)[:num_items_to_slice])
      display(list(values)[-num_items_to_slice:])

  column = df[column_name]
  unique_values = column.unique()
  try:
      sorted = np.sort(unique_values)
      print("Unique values sorted, head and tail:")
      show_head_and_tail(sorted)
  except TypeError as error:
      print(f"Could not sort values: {error}")
      print("..so let's try filtering NULL values and then sorting")
      non_null_uniques = df.loc[~df[column_name].isnull(), column_name].unique()
      sorted = np.sort(non_null_uniques)
      show_head_and_tail(sorted)


def cast_to_type(column, maybe_type):
  try:
    column.astype(maybe_type)
    print(f"Casting to {maybe_type} was successful")
  except ValueError as error:
    print(f"Could not cast to {maybe_type}: {error}")


def find_non_default_missing_values(df, column_name, maybe_type):
  long_separator_amount = 80
  short_separator_amount = 40

  print_separator("*", long_separator_amount, f"Finding non default missing values for column \"{column_name}\"")

  print(f"Column \"{column_name}\" has datatype: {df.dtypes[column_name]}")

  column = df[column_name]

  # A
  print_separator("-", short_separator_amount, "A: Looking at unique values")
  look_at_unique_values(column)

  # B
  print_separator("-", short_separator_amount, "B: Sorting and looking at the edges")
  look_at_edges(df, column_name)

  # C
  print_separator("-", short_separator_amount, f"C: Casting to type: {maybe_type}")
  cast_to_type(column, maybe_type)

  # D
  print_separator("-", short_separator_amount, "D: Looking at frequency")
  display(column.value_counts(dropna=False))

  print("\n")

find_non_default_missing_values(df_species, "Species", "string")
```

```python
# Now let's focus on cleaning the data in the species column.
# I will create a function that lists the species we want listed as a dictionary key, and lists the different spellings as values of that key.
# I also create a list with rejection keywords, which filters out the data that we do not want, e.g. data that contains 'or' or a '?', because in these sort of cases it is not certain what species caused the attack.
# The function stores the data in a new column called 'clean_species' that contains the clean data.
# I found the shark species by looking at the diagnostics above, and by checking the table. With this function, it seems most descriptions are categorized correctly.
# Instead of a dictionary, it would ofcourse also be possible to just make a list, which would be faster. However, this would not solve the issue of having different spellings for the same shark species.

species_dict = {
    'white shark': ['white shark'],
    'cookiecutter shark': ['cookiecutter shark'],
    'sevengill shark': ['7-gill shark', 'seven-gill shark', 'sevengill shark', 'broadnose sevengill shark'],
    'wobbegong shark': ['wobbegong', 'wobbegong shark'],
    'bronze whaler shark': ['bronze whaler', 'bronze whale', 'bronze whaler shark', 'whaler shark'],
    'blue pointer shark': ['blue pointer', 'blue shark', 'blue pointer shark'],
    'bluenosed shark': ['bluenose shark', 'bluenosed shark', 'blue nosed shark', 'blue nose shark', 'blue-nosed shark', 'blue whaler', 'blue shark', ],
    'sandtiger shark': ['sand shark', 'sandshark', 'sandtiger shark'],
    'hammerhead shark': ['hammerhead shark'],
    'bull shark': ['bull shark'],
    'tiger shark': ['tiger shark'],
    'lemon shark': ['lemon shark'],
    'nurse shark': ['grey nurse shark', 'tawny nurse shark', 'nurse shark'],
    'zambesi shark': ['zambesi shark', 'zambezi shark'],
    'grey reef shark': ['gray shark', 'grey shark', 'grey reef shark'],
    'mako shark': ['mako shark'],
    'blacktip reef shark': ['blacktip reef shark', 'blacktip shark', 'blacktip "reef" shark', "black tipped"],
    'reef shark': ['reef shark'],
    'oceanic whitetip shark': ['oceanic whitetip shark', 'oceanic whitetip', 'whitetip shark'],
    'basking shark': ['basking shark'],
    'copper shark': ['copper shark'],
    'cow shark': ['cow shark'],
    'port jackson shark': ['port jackson shark'],
    'salmon shark': ['salmon shark'],
    'porbeagle shark': ['porbeagle shark'],
    'galapagos shark': ['galapagos shark'],
    'raggedtooth shark': ['raggedtooth shark'],
    'caribbean reef shark': ['carribean reef shark'],
    'sandbar shark': ['sandbar shark'],
    'starry smoothound shark': ['starry smoothound shark'],
    'angel shark': ['angel shark'],
    'tresher shark': ['tresher shark'],
    'goblin shark': ['goblin shark'],
    'spinner shark': ['spinner shark'],
    'carpet shark': ['carpet shark'],
    'shovelnose shark': ['shovelnose shark'],
    'leopard shark': ['leopard shark'],
    'dusky shark': ['dusky shark'],
    'silky shark': ['silky shark'],
    'catshark': ['catshark'],
    'dogfish': ['dogfish']
    }

rejection_keywords = [
    " or ",
    "possibly",
    "probable",
    "no shark involvement",
    "shark involvement not confirmed",
    "questionable",
    "shark involvement prior to death was not confirmed",
    "shark involvement prior to death unconfirmed",
    "?",
    "unknown",
    "invalid",
    "unidentified",
    "small shark",
    "large shark"
    ]


def shark_sorter(shark_description, column = 'Species'):
  for keyword in rejection_keywords:
    if keyword in shark_description:
      return None
  for key, values in species_dict.items():
    for value in values:
      if value in shark_description:
        return key

df_species['clean_species'] = df_species['Species'].apply(func = shark_sorter).copy()

df_species.sort_values('clean_species')
```

#**QUESTION 2:**

Are children more likely to be attacked  by sharks?

Assumption:
- Adults are more likely to be attacked by sharks, because they engage more in water sports, and have more risky behavior when in the water.

```python
# Question 2: Are children more likely to be attacked  by sharks?
# Dataframe with Age (and Fatal for reference)
# Children are those below age 19.
df_age = df[['Age', 'Fatal (Y/N)']]

df_age
```

```python
# Diagnostics. There is no uniformity in the data. I will create a function that turn all values into 'Child', 'Adult' or 'None'.
find_non_default_missing_values(df_age, "Age", "string")
```

```python
#Lets start by dropping all NaN's, and turn 6.5 into 6 (as that value makes below function crash)
df_age = df_age.dropna(subset = 'Age').copy()
df_age['Age'].replace("6Â½", "6", inplace = True)
df_age['Age'].replace("2Â½", "2", inplace = True)

# Some words needs to be rejected, other words are usable. We make three lists: a rejected_category, a child_list, and an adult_list to account for words that could be included.
# There are certain rows that have multiple ages in them, e.g. "18 & 20", "5 or 4". These would be categorized under else: np.NaN now, and were thus removed.
# These rows made up around 1 percent of all data left, so I think removing these rows would not have a huge impact on the data.
# I assume terms like teens/teenagers etc. are 'Child', whereas terms like mid-20s, middle-aged etc signal that someone is adult. We can therefore still try to include this non-numeric data.
# Other data may need to be excluded, we therefore should also reject some words/phrases.

rejected_category = [
    "MAKE LINE GREEN",
    'N/A',
    np.NaN,
    " ",
    "?"
    ]

child_list = [
    'young',
    'teen',
    'Teen',
    'Teens',
    '18 months',
    '9 months']

adult_list = [
    'adult',
    '20s',
    '30s',
    '40s',
    '50s',
    '60s',
    "60's",
    'Elderly',
    '>50',
    'middle-age'
]

def age_definer(victim_age):
  victim_age = victim_age.strip() #needed because some ages had a space, and were thus not recognized as a number. Strip removes that space.
  for keyword in rejected_category:
    if str(keyword) in victim_age:
      return np.NaN
  for keyword in child_list:
    if keyword in victim_age:
      return 'Child'
  for keyword in adult_list:
    if keyword in victim_age:
      return 'Adult'
  if victim_age.isnumeric() == True:
    if float(victim_age) < 19:
      return 'Child'
    elif float(victim_age) > 18:
      return 'Adult'
  else:
    return np.NaN

df_age['Child/Adult'] = df_age['Age'].apply(func = age_definer) #apply the function to df_age

df_age = df_age.dropna(subset = 'Child/Adult') #We drop once more the empty values in child/adult.

df_age.sort_values('Child/Adult')
```

```python
#Now we can create a simple graph to see whether adults or children are more likely to be attacked.
df_age_grouped = df_age.groupby(['Child/Adult'])['Child/Adult'].count().sort_values(ascending = False)
df_age_grouped.plot(kind = 'bar', xlabel = 'Child or Adult', ylabel = 'Number of Attacks', title = 'Number of attacks on children and adults', color = 'pink')
plt.xticks(rotation = 360)
plt.grid(axis = "y")
```

##**Answer Question 2:**

- In absolute numbers, adults are more likely to be attacked by sharks.

Assumptions:

##**Observations**
For many of the values in this column, we cannot say if the attack was provoked or not. I will make a function that checks whether the value in the column is unprovoked or provoked, and else to return none. We can then drop the rows.


```python
def provoked_unknown(attack_type):
  if attack_type == 'Provoked' or attack_type == 'Unprovoked':
    return attack_type
  else:
    return None

df_shark_provoke['Provoked/Unprovoked'] = df_shark_provoke['Type'].apply(func = provoked_unknown).copy()

df_shark_provoke = df_shark_provoke.dropna(subset = 'Provoked/Unprovoked')

df_shark_provoke.sort_values('Provoked/Unprovoked')
```

```python
# Graph on the amount of provoked versus unprovoked attacks:
df_provoke_data = df_shark_provoke.groupby(['Provoked/Unprovoked'])['Provoked/Unprovoked'].count().sort_values(ascending = False)
df_provoke_data.plot(kind = 'bar', xlabel = 'Unprovoked vs Provoked', ylabel = 'Number of Attacks', title = 'Unprovoked versus Provoked Attacks', color = 'blue')
plt.xticks(rotation = 360)
plt.grid(axis = "y")
```

```python
# Now, let's check whether provoked attacks were more often fatal.
# As before, we want to clean up the "Fatal (Y/N)" column for this dataframe:
df_shark_provoke['Fatal (Y/N)'].replace(" ", "", inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace(" N", "N", inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace("M", "N", inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace("y", "Y", inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace("N ", "N", inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace(np.NaN, None, inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace('UNKNOWN', None, inplace = True)
df_shark_provoke['Fatal (Y/N)'].replace('2017', None, inplace = True)

#Dropping the None values.
df_shark_provoke.dropna(subset = "Fatal (Y/N)", inplace = True)

df_shark_provoke
```

```python
# Check if data is clean:
find_non_default_missing_values(df_shark_provoke, "Fatal (Y/N)", "string")
```

```python
# Now lets plot this. We can see that this figure does not tell us much, as unprovoked attacks are in general way more common.
df_provoke_grouped = df_shark_provoke.groupby(['Provoked/Unprovoked'])['Fatal (Y/N)'].value_counts().unstack().sort_values(by = 'Y', ascending = False)
df_provoke_grouped.plot(kind = 'bar', xlabel = 'Provoked/Unprovoked', ylabel = 'Number of Attacks', title = 'Shark Fatalities for unprovoked and provoked attacks')
plt.xticks(rotation = 360)
plt.grid(axis = "y")
```

```python
# We probably want to look at the data as percentages, as shark fatalities that are unprovoked are way more common, and this distorts the data.
# So we want to know: what percentage of provoked/unprovoked attacks were fatal.

data = [
    ['Fatal Attacks | Provoked', (len(df_shark_provoke[(df_shark_provoke['Fatal (Y/N)'] == 'Y') & (df_shark_provoke['Provoked/Unprovoked'] == 'Provoked')])) / (len(df_shark_provoke[(df_shark_provoke['Provoked/Unprovoked'] == 'Provoked')])) * 100],
    ['Fatal Attacks | Unprovoked', (len(df_shark_provoke[(df_shark_provoke['Fatal (Y/N)'] == 'Y') & (df_shark_provoke['Provoked/Unprovoked'] == 'Unprovoked')])) / (len(df_shark_provoke[(df_shark_provoke['Provoked/Unprovoked'] == 'Unprovoked')])) * 100],
    ['Non-Fatal Attacks | Provoked', (len(df_shark_provoke[(df_shark_provoke['Fatal (Y/N)'] == 'N') & (df_shark_provoke['Provoked/Unprovoked'] == 'Provoked')])) / (len(df_shark_provoke[(df_shark_provoke['Provoked/Unprovoked'] == 'Provoked')])) * 100],
    ['Non-Fatal Attacks | Unprovoked', (len(df_shark_provoke[(df_shark_provoke['Fatal (Y/N)'] == 'N') & (df_shark_provoke['Provoked/Unprovoked'] == 'Unprovoked')])) / (len(df_shark_provoke[(df_shark_provoke['Provoked/Unprovoked'] == 'Unprovoked')])) * 100]
    ]

df_percentage = pd.DataFrame(data, columns = ['Attack', 'Percentage'])

df_percentage
```

```python
# And now we can plot this again:

attack = df_percentage['Attack']
percentage = df_percentage['Percentage']

fig, ax = plt.subplots()
plt.barh(attack, percentage, color = 'blue')
plt.title("Fatality in Provoked and Unprovoked Attacks")
plt.xlabel("Percentage")
plt.ylabel("Attack and Fatality")
plt.grid()
plt.show()
```

#**Question 4**:
Are certain activities more likely to result in a shark attack?

Assumption:
- Things that happen in the water, e.g. swimming or surfing, are more likely to cause attacks than activities close to shore or in a boat.


```python
# Let's start by creating a dataframe that only includes activities, and look at the data.
# There is a lot of data that is messy, and also a lot of the data includes full sentences. Since we are looking at the main activities that cause attacks, we need to sort this data based on the type of activity.
# We need the main activities that caused attacks.
# I will therefore focus on the main activities in the list, and exclude those activities that appear only once or twice.

df_type = df[['Activity']]
find_non_default_missing_values(df_type, "Activity", "string")
```

```python
# Casting all the data in the table as 'string', since float cannot be iterated over.
df_type['Activity'] = df_type['Activity'].astype(str).copy()
```

```python
# Now we create the function
# This function includes both a dictionary and lists. The dictionary is used for activities that have different spelling, or that can be grouped together, which I did in the case of surfing, diving and boarding.

activities_dict = {
    'Surfing': ['Surfing', 'Surf Skiing', 'Surf fishing', 'Scurfing', 'surfing', 'Surf', 'surf', 'Windsurfing'],
    'Diving': ['Diving', 'diving' ,'Skindiving','Skindiving', 'skin diving', 'Skin diving', 'Pearl diving', 'Scuba diving', 'Scuba Diving', 'SCUBA diving', 'Scuba', 'Shell diving'],
    'Boarding': ['Paddling', 'Paddle Boarding', 'paddleskiing', 'paddle boarding', 'Paddle boarding', 'paddle-skiing', 'paddleboarding', 'Paddle-skiing', 'Paddle', 'boarding', 'Boarding'],
    'Sea disaster': ['Sea disaster', 'Sea Disaster'],
    'Sup': ['Sup', 'SUP'],
    'Fishing': ['Fishing', 'fishing'],
    'Boat': ['boat', 'Boat'],
    'Wreck': ['Shipwreck', 'wreck'],
    'Shark activity': ['Shark watching', 'Shark tagging', 'Shark fishing', 'Shark diving', 'Tagging sharks', 'Grabbing', 'Teasing', 'Measuring sharks'],
    'Swimming': ['Swimming', 'swimming']
    }

activities_list = [
    'Wreck',
    'Washing',
    'Walking',
    'Wading',
    'Treading water',
    'Standing',
    'Sitting',
    'Seine netting',
    'Sculling',
    'Sailing',
    'Rowing',
    'Playing',
    'Snorkeling',
    'Kayaking',
    'Canoeing',
    'Floating',
    'Bathing'
    ]

rejected_words = [
    'Unknown',
    'or',
    '?',
    'nan',
    'np.nan'
]

def activity_sorter(activities, column = 'Activity'):
  for keyword in rejected_words:
    if keyword in activities:
      return None
  for activity in activities_list:
    if activity in activities:
      return activity
  for key, values in activities_dict.items():
    for value in values:
      if value in activities:
        return key
  else:
    return None

df_type['clean_activity'] = df_type['Activity'].apply(func = activity_sorter).copy()

df_type.dropna(subset = "clean_activity", inplace = True)

df_type.sort_values('clean_activity')
```

```python
# Now we make a graph of the top 10 activities:

df_typeinfo = df_type.groupby(['clean_activity'])['clean_activity'].count().sort_values(ascending = False).head(10)
df_typeinfo.plot(kind = 'barh', xlabel = 'Number of Attacks', ylabel = 'Activity', title = 'Victim activities when shark attacks happen')
plt.grid()
plt.gca().invert_yaxis()
```

##**Answer Question 4:**

Persons who are victims of shark attacks engaged in many activities. Surfing, swimming and fishing were by far the main activities in which victims engaged before being attacked by a shark.
